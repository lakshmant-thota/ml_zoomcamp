# Question 2
Create the correlation matrix for the numerical features of your dataset. In a correlation matrix, you compute the correlation coefficient between every pair of features.

What are the two features that have the biggest correlation?

age and balance
day and campaign
day and pdays
pdays and previous

```python
df[numerical].corr(method='pearson')
```

```python
# prompt: how to extract the highest correlated pair in the above matric or a pair with highest correlation followed by the lower ones

correlation_matrix = df[numerical].corr()

# Remove diagonal part of the matrix
correlation_matrix = correlation_matrix.mask(np.equal(*np.indices(correlation_matrix.shape)))

# Find the highest correlation
max_correlation = np.nanmax(correlation_matrix)
row_max, col_max = correlation_matrix[correlation_matrix == max_correlation].stack().idxmax()

print(f"Highest correlation: {row_max} and {col_max} with correlation {max_correlation}")

# Sort the correlations in descending order and print the pairs
sorted_correlations = correlation_matrix.unstack().sort_values(ascending=False)
for (row, col), correlation in sorted_correlations.items():
    if not np.isnan(correlation):
      print(f"{row} and {col}: {correlation}")
```

# Answer 2
the answer for Q2 is pdays and previous with correlation of 0.4548

#Question 5
Let's find the least useful feature using the feature elimination technique.
Train a model with all these features (using the same parameters as in Q4).
Now exclude each feature from this set and train a model without it. Record the accuracy for each model.
For each feature, calculate the difference between the original accuracy and the accuracy without the feature.
Which of following feature has the smallest difference?

age
balance
marital
previous

```python
feature_wise_scores = []
for c in df_train.columns:
  df_train_copy = df_train.copy()
  df_val_copy = df_val.copy()
  del df_train_copy[c]
  del df_val_copy[c]
  dv = DictVectorizer(sparse=False)
  train_dicts = df_train_copy.to_dict(orient='records')
  val_dicts = df_val_copy.to_dict(orient='records')
  x_train = dv.fit_transform(train_dicts)
  x_val = dv.fit_transform(val_dicts)
  model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
  model.fit(x_train,y_train)
  y_pred = model.predict(x_val)
  score = (y_pred == y_val).mean()
  diff = actual_score - score
  feature_wise_scores.append((c,score,diff))
feature_wise_scores.sort(key = lambda x:x[2])
feature_wise_scores
```

# Answer 5 : the lwast useful features are age and day and from the options the answer is "age"

# Question 6
Now let's train a regularized logistic regression.
Let's try the following values of the parameter C: [0, 0.01, 0.1, 1, 10].
Train models using all the features as in Q4.
Calculate the accuracy on the validation dataset and round it to 3 decimal digits.
Which of these C leads to the best accuracy on the validation set?

0.01
0.1
1
10
100

```python

df_train_copy = df_train.copy()
df_val_copy = df_val.copy()
dv = DictVectorizer(sparse=False)
train_dicts = df_train_copy.to_dict(orient='records')
val_dicts = df_val_copy.to_dict(orient='records')
x_train = dv.fit_transform(train_dicts)
x_val = dv.fit_transform(val_dicts)
c_scores = []
for c in [0.01 , 0.1 , 1.0 , 10.0,100.0]:
  model = LogisticRegression(solver='liblinear', C=c, max_iter=1000, random_state=42)
  model.fit(x_train,y_train)
  y_pred = model.predict(x_val)
  score = (y_pred == y_val).mean()
  c_scores.append((c,score))
c_scores.sort(key = lambda x:x[1],reverse = True)
c_scores
```

# Answer 6
both 1 and 10 are equally performing on the model but the question was to select the least
